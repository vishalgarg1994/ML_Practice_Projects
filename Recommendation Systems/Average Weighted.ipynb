```python
import numpy as np
import pandas as pd
```

```python
credits = pd.read_csv('ratings.csv')
```

```python
credits
```

```python
movies = pd.read_csv('movies.csv')
```

```python
movies
```

```python
df = pd.merge(movies,credits,how='inner',on='movieId')
```

```python
df
```

```python
s1 = df.groupby('title')['rating'].count().sort_values(ascending=False)
s2 = df.groupby('title')['rating'].mean().sort_values(ascending=False)
```

```python
df1 = pd.DataFrame()
df1['num_ratings'] = s1
df1['Title']  = df1.index
df2 = pd.DataFrame()
df2['avg_ratings'] = s2
df2['Title']  = df1.index
```

```python
df1.reset_index(drop=True, inplace=True)
df2.reset_index(drop=True, inplace=True)
```

```python
df3 = pd.merge(df1,df2,how='inner',on='Title')
```

```python
df3.num_ratings = df3.num_ratings.astype('int64')
```

```python
df3 = df3.loc[df3.num_ratings > 200]
```
