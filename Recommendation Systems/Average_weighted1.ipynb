```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
```

```python
movies.head()
```

```python
credits.head()
```

```python
movies = movies.rename(index=str,columns={"id":"movie_id"})
```

```python
df = movies.merge(credits,on='movie_id')
```

```python
df
```

```python
df = df.drop(columns=['homepage', 'title_x', 'title_y', 'status','production_countries'])
```

```python
df
```

```python
# Now we will use a formula to calculate weighted average
# W = (Rv+Cm)/v+m
#R ---> vote_average
#C ---> Avg of vote_average
#v ---> vote count
#m ---> 

v=df['vote_count']
R=df['vote_average']
C=df['vote_average'].mean()
m=df['vote_count'].quantile(0.70) # vote count at 70% quantile
R
```

```python
df['weighted_average'] = ((R*v)+(C*m))/(v+m)
```

```python
df['weighted_average'].sort_values(ascending=False)
```

```python
new_df = df.sort_values('weighted_average',ascending=False)
new_df[['original_title', 'vote_count', 'vote_average', 'weighted_average', 'popularity']].head()    
```

```python
new_df
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x=new_df['weighted_average'].head(10), y=new_df['original_title'].head(10), data=new_df)
```

```python
new_df1 = df.sort_values('popularity',ascending=False)
plt.figure(figsize=(12,6))
sns.barplot(x=new_df1['popularity'].head(10), y=new_df1['original_title'].head(10), data=new_df1)
```

```python
# Now we can take priority as 50-50 for both popularity and weighted_average
from sklearn.preprocessing import MinMaxScaler
sc = MinMaxScaler()
scaled_df = sc.fit_transform(df[['weighted_average','popularity']])
scaled_df = pd.DataFrame(scaled_df,columns = ['weighted_average','popularity'] )
```

```python
df[['weighted_average','popularity']] = scaled_df
```

```python
df['score'] = df['weighted_average']*0.5 + df['popularity']*0.5
```

```python
new_df2 = df.sort_values('score',ascending=False)
plt.figure(figsize=(12,6))
sns.barplot(x=new_df2['score'].head(10), y=new_df2['original_title'].head(10), data=new_df2)
```
